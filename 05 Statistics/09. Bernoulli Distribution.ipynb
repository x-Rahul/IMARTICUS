# Bernoulli Distribution

The Bernoulli distribution is a discrete probability distribution for a random variable which takes the value 1 with probability $ p $ and the value 0 with probability $ 1 - p $. It is named after the Swiss mathematician Jacob Bernoulli. This distribution is used to model a single trial of a process that can result in one of two outcomes, often referred to as "success" and "failure."

### Key Characteristics of the Bernoulli Distribution:

1. **Support:** The random variable $ X $ can take values in $ \{0, 1\} $.

2. **Probability Mass Function (PMF):**
   $$
   P(X = x) = \begin{cases} 
   p & \text{if } x = 1 \\
   1 - p & \text{if } x = 0 
   \end{cases}
   $$
   where $ 0 \leq p \leq 1 $.

3. **Mean (Expected Value):**
   $$
   E[X] = p
   $$

4. **Variance:**
   $$
   \text{Var}(X) = p(1 - p)
   $$

5. **Skewness:**
   $$
   \frac{1 - 2p}{\sqrt{p(1 - p)}}
   $$

6. **Kurtosis:**
   $$
   \frac{1 - 6p(1 - p)}{p(1 - p)}
   $$

### Applications:
- **Modeling binary outcomes:** Used to model any experiment that has exactly two possible outcomes, such as flipping a coin (heads/tails), success/failure in a trial, yes/no responses, etc.
- **Building blocks for other distributions:** Serves as the foundation for the Binomial distribution, which is the distribution of the sum of independent Bernoulli trials.

## Example

Consider a scenario where you flip a fair coin. The probability of getting heads (success) is $ p = 0.5 $, and the probability of getting tails (failure) is $ 1 - p = 0.5 $.

1. **Probability Mass Function (PMF):**
   $$
   P(X = x) = \begin{cases} 
   0.5 & \text{if } x = 1 \text{ (heads)} \\
   0.5 & \text{if } x = 0 \text{ (tails)} 
   \end{cases}
   $$

2. **Mean (Expected Value):**
   $$
   E[X] = 0.5
   $$

3. **Variance:**
   $$
   \text{Var}(X) = 0.5 \times (1 - 0.5) = 0.25
   $$

4. **Skewness:**
   $$
   \frac{1 - 2 \times 0.5}{\sqrt{0.5 \times (1 - 0.5)}} = 0
   $$

5. **Kurtosis:**
   $$
   \frac{1 - 6 \times 0.5 \times (1 - 0.5)}{0.5 \times (1 - 0.5)} = -2
   $$

This example demonstrates the basic properties and calculations for a Bernoulli distribution with $ p = 0.5 $.


### Bernoulli Distribution Example

**Example:**
Suppose the probability of getting heads in a single coin toss is 0.5.

#### Calculations:
- Probability of success: $p = 0.5$

#### Probability Mass Function (PMF):
The PMF of a Bernoulli distribution is given by:

$$ P(X = k) = p^k (1 - p)^{1 - k} \quad \text{for } k \in \{0, 1\} $$

#### Plot:
```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Parameters
p = 0.5

# Bernoulli distribution
x = np.arange(0, 2)
pmf = bernoulli.pmf(x, p)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(x, pmf, color='blue', alpha=0.7)
plt.title('Bernoulli Distribution (p=0.5)')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.xticks(x, ['0 (Failure)', '1 (Success)'])
plt.grid(True)
plt.show()
