# Advertising Dataset

Our goal is to develop an accurate model that can be used to predict sales on the basis of the three media budgets.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
Advertising = pd.read_csv("Dataset/Advertising.csv")
Advertising.head()
```

```python
Advertising.info()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.scatterplot(ax=axes[0], data=Advertising, x='TV', y='sales', color='r')
sns.scatterplot(ax=axes[1], data=Advertising, x='radio', y='sales', color='g')
sns.scatterplot(ax=axes[2], data=Advertising, x='newspaper', y='sales', color='b')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.regplot(ax=axes[0], data=Advertising, x='TV', y='sales', scatter_kws={'color': 'r'}, line_kws={'color':'k'})
sns.regplot(ax=axes[1], data=Advertising, x='radio', y='sales', scatter_kws={'color':'g'}, line_kws={'color': 'k'})
sns.regplot(ax=axes[2], data=Advertising, x='newspaper', y='sales', scatter_kws={'color':'b'}, line_kws={'color': 'k'})

plt.tight_layout()
plt.show()
```
