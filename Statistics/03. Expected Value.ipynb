### Expected Value $ E(X) $

The expected value is the probability-weighted sum of all possible values that the random variable can take. The expected value, denoted as $ E(X) $, of a random variable $ X $ is a measure of the central tendency of the possible values that $ X $ can take. It is essentially the long-run average value of $ X $ over numerous repetitions of the experiment or process that generates $ X $. The expected value is often referred to as the mean of the random variable.

#### Expected Value for Discrete Random Variables

For a discrete random variable $ X $ with possible values $ x_1, x_2, \ldots, x_n $ and corresponding probabilities $ P(X = x_i) $, the expected value is calculated as:

$$
E(X) = \sum_{i} x_i \cdot P(X = x_i)
$$

This formula sums up each possible value of $ X $, weighted by its probability.

For a binomial random variable $ X $, which represents the number of successes in $ n $ independent trials with a success probability $ p $, the probability of exactly $ k $ successes is given by:

$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
$$

where $ \binom{n}{k} $ is the binomial coefficient, representing the number of ways to choose $ k $ successes out of $ n $ trials.

The binomial coefficient $ \binom{n}{k} $ is calculated as:

$$
\binom{n}{k} = \frac{n!}{k!(n-k)!}
$$

The expected value of a binomial random variable $ X $ is then:

$$
E(X) = n \cdot p
$$

#### Expected Value for Continuous Random Variables

For a continuous random variable $ X $ with probability density function (PDF) $ f(x) $, the expected value is calculated as:

$$
E(X) = \int_{-\infty}^{\infty} x \cdot f(x) \, dx
$$

This formula integrates over all possible values of $ X $, weighted by the density function $ f(x) $.

#### Properties of Expected Value

1. **Linearity**: The expected value operator is linear, meaning for any random variables $ X $ and $ Y $ and constants $ a $ and $ b $,

   $$
   E(aX + bY) = aE(X) + bE(Y)
   $$

2. **Expectation of a Constant**: If $ c $ is a constant, then

   $$
   E(c) = c
   $$

3. **Expectation of a Sum**: For random variables $ X_1, X_2, \ldots, X_n $,

   $$
   E\left(\sum_{i=1}^{n} X_i\right) = \sum_{i=1}^{n} E(X_i)
   $$

#### Examples

1. **Discrete Example**: Suppose $ X $ represents the result of a single roll of a fair six-sided die. The possible values of $ X $ are $ 1, 2, 3, 4, 5, $ and $ 6 $, each with a probability of $ \frac{1}{6} $. The expected value is:

   $$
   E(X) = \sum_{i=1}^{6} i \cdot \frac{1}{6} = \frac{1}{6}(1 + 2 + 3 + 4 + 5 + 6) = \frac{21}{6} = 3.5
   $$

2. **Continuous Example**: Suppose $ X $ is a continuous random variable uniformly distributed between 0 and 1. The PDF is $ f(x) = 1 $ for $ 0 \leq x \leq 1 $ and 0 otherwise. The expected value is:

   $$
   E(X) = \int_{0}^{1} x \cdot 1 \, dx = \left[ \frac{x^2}{2} \right]_{0}^{1} = \frac{1}{2}
   $$
