```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt # for better visualization
import seaborn as sns # for better visualization
```

```python
x1=[1,2,5,8,10,23]
x2=list(map(lambda x : x**2, x1)) # by using lambda and map , we are getting new list with squared number
x3=[3,8,10,15,14,20]
x4=[8,12,14,18,20,23]
```

```python
# plt.plot_name(arguments1, arg2, ....)
# plt.title("A Vs B")
# plt.xlabel("What to display on x axis")
# plt.ylabel("What to display on y axis")
# plt.show() # display the plot
```

```python
plt.figure(figsize=(7,3))
plt.plot(x1, x2)  # Line chart
plt.title("Line Chart btw x1 and x2")
plt.xlabel("Displaying x1 value")
plt.ylabel("Diplaying Square of x1 elements (x2)")
plt.show()
```

```python
plt.figure(figsize=(7,3))
plt.plot(x1, x2, color="red",linestyle="-.", linewidth=2, marker="h",
         markersize=10, markerfacecolor="black",markeredgecolor="blue")  # Line chart
plt.title("Line Chart btw x1 and x2")
plt.xlabel("Displaying x1 value")
plt.ylabel("Diplaying Square of x1 elements (x2)")
plt.tick_params(labelsize=10,color="red", labelcolor="black")  # changing the axis label font
plt.xlim(0,60)
# plt.ylim
plt.xticks([0,20,40], ["A", "B", "C"])
# plt.yticks
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7,3))
plt.plot(x1, x2, color="red",linestyle="-.", linewidth=2, marker="h",
         markersize=10, markerfacecolor="black",markeredgecolor="blue", 
         label="First Line X1-X2")  # Line chart

plt.plot(x3, x4, color="blue",linestyle="-", linewidth=2, marker="o",
         markersize=6, markerfacecolor="orange",markeredgecolor="blue",
        label="Second Line X3-X4")  # Line chart

plt.title("Line Chart btw x1 and x2")
plt.xlabel("Displaying x1 value")
plt.ylabel("Diplaying Square of x1 elements (x2)")
plt.tick_params(labelsize=10,color="red", labelcolor="black")  # changing the axis label font
plt.xlim(0,60)
# plt.ylim
plt.xticks([0,20,40], ["A", "B", "C"])
# plt.yticks
plt.grid(True) 
plt.legend()
plt.show()
```

```python
a=np.linspace(-2*np.pi, 2*np.pi, 2560, endpoint=True)
s=np.sin(a)
c=np.cos(a)
plt.plot(c, label="cosine")
plt.plot(s, label="sine")
plt.legend()
plt.show()
```
