# Plots

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
x1=[1, 2, 5, 8, 10, 23]
x2 = list(map(lambda x:x**2, x1))
x3 = [3, 8, 10, 15, 14, 20]
x4 = [8, 12, 14, 18, 20, 23]
```

```python
plt.plot(x1, x2)
```

```python
plt.figure(figsize=(6,3))
plt.plot(x1, x2)
plt.title("Line Chart")
plt.xlabel("x1 data")
plt.ylabel("x2 data (sq of x)")
```

```python
plt.figure(figsize=(6,3))
plt.plot(x1, x2, 
         color='c', 
         linestyle='-.', 
         linewidth = 2, 
         marker='p', 
         markersize=8,
         markerfacecolor = 'b',
         markeredgecolor = 'k'
)
plt.title("Line Chart")
plt.xlabel("x1 data")
plt.ylabel("x2 data (sq of x1)")
plt.xlim(0, 30)
plt.xticks(color='g')
# plt.xticks([0, 10, 30], ["A", "B", "C"])
plt.tick_params(labelsize=12, color = 'r')
```

```python
plt.figure(figsize=(6,3))
plt.plot(x1, x2, 
         color='c', 
         linestyle='-.', 
         linewidth = 2, 
         marker='p', 
         markersize=8,
         markerfacecolor = 'b',
         markeredgecolor = 'k'
)
plt.plot(x3, x4, 
         color='r', 
         linestyle='-.', 
         linewidth = 2,
         marker='*', 
         markersize=8,
         markerfacecolor = 'b',
         markeredgecolor = 'k'
)
plt.title("Line Chart")
plt.xlabel("x1 data")
plt.ylabel("x2 data (sq of x1)")
plt.xlim(0, 30)
plt.xticks(color='g')
# plt.xticks([0, 10, 30], ["A", "B", "C"])
plt.tick_params(labelsize=12, color = 'r')
```

```python
theta = np.linspace(-2*np.pi, 2*np.pi, 2560, endpoint=True)
s = np.sin(theta)
c = np.cos(theta)
plt.plot(s, color='g')
plt.plot(c, color='r')
```

## Bar chart

```python
superstore = pd.read_excel(r"/Users/rahulmeena/Imarticus/GitHub/IMARTICUS/Datasets/class_exercise/Global Superstore.xls")
```

```python
superstore.head()
```

```python
d = pd.DataFrame(superstore['Segment'].value_counts()).reset_index()
d
```

```python
plt.bar(x=d["Segment"], height=d["count"], color=['r', 'g', 'b'])
```

```python
sns.barplot(x=d["Segment"], y=d["count"])
```

```python
plt.pie(x=d["count"],labels= d["Segment"], colors=['r','g','b'])
plt.show()
```

```python
plt.pie(x=d["count"],labels= d["Segment"], 
        colors=['r','g','c'],
        shadow="True",
        explode=[0,0,.2],
        autopct="%.2f%%",
        textprops={"fontsize": 12, "color": "navy"},
        wedgeprops={"linewidth": 3, "edgecolor":"black"},
        pctdistance=.5
        
)
plt.show()
```

```python
car_sales =  pd.read_csv(r"/Users/rahulmeena/Imarticus/GitHub/IMARTICUS/Datasets/class_exercise/Car_sales.csv")
```

```python
car_sales.head()
```

```python
sns.histplot(data=car_sales)
plt.show()
```

```python
sns.histplot(data=car_sales, x='Sales_in_thousands', 
             hue='Vehicle_type',
             multiple="dodge"
)
```

```python
plt.figure(figsize=(9,4))
sns.displot(data=car_sales, x="Sales_in_thousands",
            hue="Vehicle_type",
            kind="kde")
plt.show()
```

```python
plt.figure(figsize=(9,4))
sns.displot(data=car_sales, x="Sales_in_thousands",
            hue="Vehicle_type",
            multiple="stack",
            kind="kde")
plt.show()
```

```python
sns.jointplot(data=car_sales, x="Sales_in_thousands", y="Price_in_thousands",
            hue="Vehicle_type",
            kind="kde")
plt.show()
```
